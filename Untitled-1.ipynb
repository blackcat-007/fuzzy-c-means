```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs

def calculate_membership(data, centroids, k, m):
    # Calculate new membership matrix U_new based on centroids and data points
    U_new = np.zeros((data.shape[0], k))
    for i in range(k):    
        U_new[:, i] = np.linalg.norm(data - centroids[i, :], axis=1)
    U_new = 1 / (U_new ** (2 / (m - 1)) * np.sum((1 / U_new) ** (2 / (m - 1)), axis=1)[:, np.newaxis])
    return U_new
def calculate_centeroids(data,k,u,m):    
    centeroids=np.zeros((k,data.shape[1]))
    for i in range(k):
        centeroids[i,:]=np.sum((u[:,i]**m)[:,np.newaxis]*data,axis=0)/np.sum(u[:,i]**m)
    return centeroids


# Step 2: Initialize parameters for Fuzzy C-Means
K = 2  # Number of clusters
m = 1.5  # Fuzziness parameter
max_iter = 100  # Maximum iterations

# Initialize membership matrix U with random probabilities
U = np.random.rand(n_samples, K)
U = U / np.sum(U, axis=1, keepdims=True)  # Normalize so each row sums to 1

# FCM algorithm
for iteration in range(max_iter):
    # Step 3: Calculate centroids
    centroids = calculate_centeroids(X,K,U,m)

    # Step 4: Update membership matrix U
    U_new = calculate_membership(X, centroids, K, m)
    # Check for convergence
    if np.linalg.norm(U_new - U) < 0.00001:
        break
    U = U_new

# Plot the final results
colors = []
for i in range(n_samples):
    if U[i, 0] > 0.95:  # Points totally in cluster 1
        colors.append("green")
    elif U[i, 1] > 0.95:  # Points totally in cluster 2
        colors.append("yellow")
    elif U[i, 0] > U[i, 1]:  # Points leaning towards cluster 1
        colors.append("blue")
    else:  # Points leaning towards cluster 2
        colors.append("orange")

# Plot using Seaborn
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=colors, palette={"green": "green", "yellow": "yellow", "blue": "blue", "orange": "orange"}, legend=False)
plt.scatter(centroids[:, 0], centroids[:, 1], color="red", marker="X", s=100, label="Centroids")
plt.title("Fuzzy C-Means Clustering")
plt.show()

```
